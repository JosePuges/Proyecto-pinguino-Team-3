└── **📌 Descripción de ramas**

**🟢 `main`**
- Rama principal del proyecto
- Código estable y final

**🟡 `develop`**
- Rama de integración
- Aquí se juntan todas las features antes de pasar a main

**🔵 `feature/*`**
- Ramas de trabajo individual
- Cada una contiene una funcionalidad específica:
  - `feature/limpieza-datos` → limpieza y preparación
  - `feature/analisis-exploratorio` → EDA
  - `feature/visualizaciones` → gráficos
  - `feature/modelo` → modelos de ML

**🔄 Flujo de trabajo**

1. Crear rama desde `develop`
2. Trabajar en la feature
3. Hacer commit y push
4. Pull request hacia `develop`
5. Merge cuando esté validado

└── **COMMITS**

**FLUJO DE TRABAJO**

| Orden | Acción | Comando / Procedimiento | Notas de Seguridad |
| :--- | :--- | :--- | :--- |
| 1 | Preparación | git clone <url> | Solo la primera vez para bajar el proyecto. |
| 2 | Aislamiento | python -m venv venv | Cada uno crea su entorno virtual localmente. |
| 3 | Configuración | modificar .gitignore | Asegurarse de incluir /venv y .ipynb_checkpoints. |
| 4 | Salto a Rama | git switch <rama_asignada> | Te mueves a la rama específica ya creada. |
| 5 | Sincronización | git pull origin dev | Crítico: Traer los cambios de otros antes de empezar. |
| 6 | Desarrollo | Trabajo en el Notebook |  |
| 7 | Comitear | git add . + git commit -m "..." | Guardar cambios con Conventional Commits. |
| 8 | Publicación | git push| Subir los cambios a la nube (GitHub/GitLab). |
| 9 | Integración | Abrir Pull Request (PR) | Se solicita unir tu rama con la rama dev. |
| 10 | Limpieza | git switch dev + git branch -d <rama> | Solo borrar la rama local tras confirmar el merge. Esta opción es más fácil desde remoto |


**PROPOSITOS**
| Tipo | Propósito | Ejemplo de uso |
| :--- | :--- | :--- |
| feat | Una nueva característica o funcionalidad. | feat: add database connection |
| fix | Corrección de un error o bug. | fix: resolve null pointer exception |
| docs | Cambios solo en la documentación. | docs: update installation instructions |
| style | Cambios de formato (espacios, comas) que no afectan al código. | style: fix indentation in main.py |
| refactor | Mejora del código que no arregla un bug ni añade una función. | refactor: simplify data processing loop |
| perf | Un cambio de código que mejora el rendimiento. | perf: reduce memory usage in training |
| test | Añadir pruebas faltantes o corregir pruebas existentes. | test: add unit tests for login module |
| chore | Tareas de mantenimiento, dependencias o herramientas. | chore: update dependencies in package.json |
| build | Cambios que afectan al sistema de compilación o dependencias externas. | build: upgrade gradle version |
