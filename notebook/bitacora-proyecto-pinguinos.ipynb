## **CONFLICTO ⚠️ *Resolución conflicto merge***

##### ¿Qué pasó?

Al hacer `git pull`, Git detectó que dos personas habían modificado el mismo archivo (`proyecto-curie-pinguinos.ipynb`) y no pudo fusionarlos automáticamente:

*CONFLICT (content): Merge conflict in proyecto-curie-pinguinos.ipynb
Automatic merge failed; fix conflicts and then commit the result.*

> No es un error como tal — es Git avisando de que necesita intervención humana.


##### ¿Cómo se resolvió?

1. `git add .` — se marcaron los conflictos como resueltos
2. `git commit -m "guardado"` — se confirmó el merge con commit




**`git reset --hard origin/dev`**

Ese comando hace tres cosas importantes:

`git reset` — mueve el puntero de tu rama local   
`--hard —` descarta todos los cambios locales sin preguntar (staged y unstaged)   
`origin/dev` — al estado exacto en que está la rama dev en el remoto   

☢️ **Es destructivo:** cualquier cambio local que no hayas pusheado se pierde para siempre.
Se usa típicamente cuando quieres descartar todo lo local y sincronizarte completamente con el remoto.

# **Problema - Solución** que nos encotramos al estar trabajando en una rama incorrecta:

Hemos trabajado sin querer en la rama `dev`, cuando deberíamos haberlo hecho en la rama `6-analisis-univariado`.    
Para cambiar de rama **sin perder los cambios**, desde `dev` hacemos:

```bash
git checkout 6-analisis-univariado
```

Esto nos lleva directamente a la rama destino. Desde ahí hacemos:

```bash
git merge dev
```

> Si la rama no existiese, podríamos crearla y cambiar a ella en un solo paso con `git checkout -b nombre-nueva-rama`.

```
📁 Proyecto Pingüinos
└── 📁 app
    ├── 📁 content
    │   ├── 📄 00_home.py
    │   ├── 📄 01_inspeccion.py
    │   ├── 📄 02_numericas.py
    │   ├── 📄 03_categoricas.py
    │   ├── 📄 04_relaciones.py
    │   └── 📄 05_dashboard.py
    ├── 📁 pages
    │   ├── 📄 1_inspeccion.py
    │   ├── 📄 2_variables_numericas.py
    │   ├── 📄 3_variables_categoricas.py
    │   ├── 📄 4_relaciones.py
    │   └── 📄 5_dashboard_final.py
    ├── 📁 utilidades
    │   ├── 📄 carga_datos.py
    │   ├── 📄 contenido.py
    │   ├── 📄 graficas.py
    │   ├── 📄 inspeccion.py
    │   ├── 📄 limpieza.py
    │   ├── 📄 nombres.py
    │   ├── 📄 ui.py
    │   └── 📄 inicio.py
    └── 📁 notebook
        ├── 📓 bitacora.ipynb
        └── 📓 proyecto_curie_pinguinos.ipynb

# <h3><span style="background-color: #8e44ad; padding: 2px 8px; border-radius: 4px; color: white;">📅 **23/04/2026**</span></h3>

## **Del Notebook al Módulo**

El punto de partida fue un notebook de análisis exploratorio (EDA) donde
desarrollamos el código de forma lineal: cargábamos datos, los limpiábamos,
generábamos gráficas y extraíamos conclusiones. Este flujo es ideal para
explorar, pero no para compartir ni reutilizar.

El primer gran paso fue **refactorizar**: identificar bloques de código
repetidos o con una responsabilidad clara y convertirlos en **funciones
reutilizables**, organizadas en módulos dentro de la carpeta `utilidades/`.

| Módulo | Responsabilidad |
|---|---|
| `carga_datos.py` | Leer y cachear el dataset |
| `limpieza.py` | Transformaciones y valores nulos |
| `graficas.py` | Generación de visualizaciones |
| `inspeccion.py` | Estadísticos y estructura del df |
| `ui.py` | Componentes visuales de la interfaz |
| `nombres.py` | Constantes y etiquetas reutilizables |

---

### ¿Qué es Streamlit?

**Streamlit** es un framework de Python que permite convertir scripts en
aplicaciones web interactivas.

Con Streamlit, cada vez que el usuario interactúa con un widget (un slider,
un filtro, un botón), el script se re-ejecuta de arriba a abajo, actualizando
la interfaz automáticamente.

Ejecutando `streamlit run app.py` en la terminal, Streamlit levanta un
**servidor local** accesible desde el navegador en `http://localhost:8501`.

---

### ¿Qué es un Localhost?

Un **localhost** es un servidor que corre en tu propia máquina. En lugar de
subir la app a internet, el servidor vive en tu ordenador y el navegador se
conecta a él localmente. Es el entorno ideal para desarrollar y testear antes
de desplegar.

---

### Arquitectura final: Pages + Content + Utilidades

Para estructurar la app de forma escalable, separamos tres capas:

- **`pages/`** — Cada archivo es una página del menú lateral de Streamlit.
  Solo orquestan: llaman a funciones, no tienen lógica propia.
- **`content/`** — Contiene el "guion" de cada página: qué se muestra,
  en qué orden y con qué texto.
- **`utilidades/`** — La capa de lógica: carga de datos, limpieza,
  gráficas y componentes UI reutilizables.
Usuario
└── Streamlit (pages/)
└── Content (qué mostrar)
└── Utilidades (cómo hacerlo)
└── Dataset Palmer Penguins

Esta separación sigue el principio de **responsabilidad única**: cada capa
sabe hacer una sola cosa, lo que hace el proyecto más fácil de mantener,
depurar y ampliar.

---

### ¿Y Stitch?

**Stitch** (o el patrón *stitching*) es la práctica de **coser** distintos
módulos o fuentes de datos en un único flujo coherente. En nuestro proyecto,
`inicio.py` actúa como punto de entrada que une la carga de datos, la
configuración de la interfaz y el enrutamiento entre páginas, asegurando que
todos los módulos hablen entre sí de forma ordenada.

---

> 💡 **Conclusión:** Pasamos de un análisis lineal en un notebook a una
> aplicación web modular, mantenible y navegable, aplicando buenas prácticas
> de ingeniería de software: funciones reutilizables, separación de
> responsabilidades y un framework que convierte Python puro en una interfaz
> web interactiva.

# <h3><span style="background-color: #8e44ad; padding: 2px 8px; border-radius: 4px; color: white;">📅 **24/04/2026**</span></h3>

## **🎨Frontend: Poniéndolo Bonito**

Una vez que la lógica funcionaba, llegó el momento de que la app
*también tuviera buena pinta*. Porque los datos pueden ser muy buenos,
pero si la presentación no acompaña, el mensaje se pierde.

### El reto

Streamlit te da una interfaz funcional casi sin esfuerzo, pero
conseguir que se vea **exactamente como quieres** es otra historia.
Nos peleamos con los márgenes, con los colores, con los componentes
que no se comportaban como esperábamos.

### ¿Cómo lo resolvimos?

Aquí es donde entró **Stitch**. Stitch es una herramienta de diseño
que nos permitió visualizar y ajustar la interfaz de forma más
intuitiva, para luego trasladar esos estilos a la app. En lugar de
ir a ciegas probando valores de CSS, pudimos ver el resultado antes
de escribir una sola línea de código.

> 💡 Una buena interfaz no es decoración — es parte del análisis.
> Si el usuario no entiende lo que ve, los datos no sirven de nada.
