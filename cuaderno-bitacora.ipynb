 
### Observaciones
>  **Valores Nulos**    

No hay una regla universal de "siempre imputar" o "siempre borrar". Con nulos se decide por causa, porcentaje e impacto. Si representan menos del 5%, pueden eliminarse siempre que se justifique la decisión. En caso de no hacerlo, también existe la opción de imputarlos.   

| Paso | Que revisar | Regla orientativa | Decisión sugerida |
|---|---|---|---|
| 1 | Tipo de nulo | MCAR / MAR / MNAR | Identificar origen antes de actuar |
| 2 | Volumen de nulos | % por columna, fila, segmento y tiempo | Medir primero, decidir después |
| 3 | Importancia de variable | Critica vs secundaria | Priorizar calidad en variables criticas |
| 4 | Nulos bajos | Menos de 5% y sin patrón | Imputación simple (mediana/moda) o mantener |
| 5 | Nulos medios | Entre 5% y 30% | Imputación por grupos/modelo + flag de faltante |
| 6 | Nulos altos | Mas de 30%-40% | Considerar eliminar variable si no es critica |
| 7 | Nulo estructural | "No aplica" | No imputar; tratar como categoría informativa |
| 8 | Trazabilidad | Método y motivo | Documentar decisión y validar impacto |

**PROTOCOLO NULOS (GENERICO)**
1) Entiende el origen: MCAR, MAR o MNAR.
2) Mide % de nulos por columna/fila/segmento/tiempo.
3) Si <5%: imputacion simple o mantener.
4) Si 5-30%: imputacion cuidada + flag "faltante=1".
5) Si >30-40%: valorar eliminar variable no critica.
6) Si es "no aplica": no imputar, categoría aparte.
7) Compara resultados con y sin imputación.
8) Documenta que hiciste, por que y el riesgo.

**MCAR, MAR, MNAR**
MCAR (Missing Completely At Random): faltan datos al azar, sin relación con ninguna variable del dataset.
MAR (Missing At Random): la falta de datos se explica por otras variables observadas (ejemplo: falta ingreso según rango de edad).
MNAR (Missing Not At Random): la falta depende del propio valor faltante o de una causa no observada (ejemplo: no declarar ingresos cuando son muy altos o muy bajos).
